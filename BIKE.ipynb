```
import seaborn as sns
columns = ['temp','atemp', 'windspeed', 'casual']
sns.pairplot(train[columns])
```

```
for i, value in train[train['instant'].isna()].iterrows():
  train['instant'][i] = i+1

train.instant.isnull().any()
```

```
train['hum'].fillna(train['hum'].mean(), inplace=True)
train['cnt'].fillna(train['cnt'].mean(), inplace=True)
train['casual'].fillna(train['casual'].mean(), inplace=True)

Y = train['cnt']
X = train[['hum','casual']]
```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

model = LinearRegression()
model.fit(X,Y)

y_pred = model.predict(X)
r2_score(Y, y_pred)
```
